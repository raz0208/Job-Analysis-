# **Data Analyst Job Analysis**

The data used for this notebook is from LinkedIn jobs page. Only Key Responsibilities and job Requirments are slected as main data to this analys.

```python
# Import Required Libraries
import pandas as pd
import re
```

```python
# Read And Read The file
with open('DA-Skills.txt', 'r', encoding='utf-8') as f:
    content = f.read()

# Show some of the data in file
print(content[:500])
```

```python
# Split the content into jobs using the pattern like "1)", "2)", etc as objects.
# We look for a number followed by a parenthesis at the start of a line or after a newline.
jobs_raw = re.split(r'\n(?=\d+\))|^(?=\d+\))', content)
jobs = [j.strip() for j in jobs_raw if j.strip()]

# Clean job IDs if necessary (ensure we have 43)
print(f"Number of jobs found: {len(jobs)}")
```

```python
# Initialize the data list
data = []

for i, job_text in enumerate(jobs):
    row = {'Job Description': f"Job {i+1}"}
    for skill, patterns in skill_keywords.items():
        found = 0
        for pattern in patterns:
            if re.search(pattern, job_text, re.IGNORECASE):
                found = 1
                break
        row[skill] = found
    data.append(row)

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('skill_frequency_matrix.csv', index=False)

df.head()
```

```python
# Drop columns where the frequency of "1" is less than 5
df_cleaned = df.drop(columns=[col for col in df.columns if col != 'Job Description' and df[col].sum() < 5])

# df_cleaned.head()

# Drop all rows where all columns are "0"
df_cleaned = df_cleaned.loc[~(df_cleaned.iloc[:, 1:] == 0).all(axis=1)]

df_cleaned
```
